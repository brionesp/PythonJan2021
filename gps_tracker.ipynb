```python
import folium
import csv
gpsTrackData = open("gps_track.txt")

csvReader = csv.reader(gpsTrackData)
# print(csvReader)
# Return the next row of the reader objects iterable elements as a list
header = csvReader.__next__() # Returns the header elements as a list
# print(header)
latIndex = header.index("lat")
longIndex = header.index("long")
# print(latIndex, longIndex)
coordList = []
# Determine the initial starting point for the map
mapa = folium.Map(location=[40.78966141, -77.85948515], zoom_start=12)
for row in csvReader:
    lat = row[latIndex]
    lon = row[longIndex]
#     coordList.append([lat, lon])
    folium.Marker([lat, lon]).add_to(mapa)
# print(coordList)
mapa

```
